```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import pandas as pd
import os
```

```python
temps = []

dF = 0.05
dT = 0.05

F_ = np.arange(-1.,2.5,dF) 
T_ = np.arange(-0.3,1.2,dT)
T, F = np.meshgrid(T_,F_)
heatmap = np.zeros_like(T)

for filename in os.listdir("./"):
    if not ".csv" in filename: continue
    print("Reading "+filename,end="\n")
    forcings = []
    filedata = pd.read_csv(filename)
    years = np.array(filedata["year"])
    temp = np.array(filedata["temp"])
    
    for forcing_num in range(1000):
        print(forcing_num,end=", ")
        forcing = np.array(filedata["forcing_"+str(forcing_num)])
        for i in range(F_.size):
            for j in range(T_.size):
                heatmap[i,j] += np.any(
                    np.logical_and(
                        np.logical_and(temp>T_[j]-dT/2.,temp<=T_[j]+dT/2.),
                        np.logical_and(forcing>F_[i]-dF/2.,forcing<=F_[i]+dF/2.)
                    )
                )
```

```python
filename = "FAR_EBM_temps_and_forcing.xlsx"
filedata = pd.read_excel(filename)
```

```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

heatmap = np.ma.masked_array(heatmap,heatmap==0.)

cmap = plt.get_cmap('Blues')
new_cmap = truncate_colormap(cmap, 0., 0.9)
new_cmap.set_bad(color = 'w', alpha = 1.)
```
