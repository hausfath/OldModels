```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import pandas as pd
import os
```

```python
temps = []

dF = 0.05
dT = 0.05

F_ = np.arange(-1.,2.5,dF) 
T_ = np.arange(-0.3,1.2,dT)
T, F = np.meshgrid(T_,F_)
heatmap = np.zeros_like(T)

for filename in os.listdir("./"):
    if not ".csv" in filename: continue
    print("Reading "+filename,end="\n")
    forcings = []
    filedata = pd.read_csv(filename)
    years = np.array(filedata["year"])
    temp = np.array(filedata["temp"])
    
    for forcing_num in range(1000):
        print(forcing_num,end=", ")
        forcing = np.array(filedata["forcing_"+str(forcing_num)])
        for i in range(F_.size):
            for j in range(T_.size):
                heatmap[i,j] += np.any(
                    np.logical_and(
                        np.logical_and(temp>T_[j]-dT/2.,temp<=T_[j]+dT/2.),
                        np.logical_and(forcing>F_[i]-dF/2.,forcing<=F_[i]+dF/2.)
                    )
                )
```

```python
filename = "Hansen_88_plot_data.xls"
filedata = pd.read_excel(filename)
```

```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

heatmap = np.ma.masked_array(heatmap,heatmap==0.)

cmap = plt.get_cmap('Blues')
new_cmap = truncate_colormap(cmap, 0., 0.9)
new_cmap.set_bad(color = 'w', alpha = 1.)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.plot([1988,1988],[-1.,2.],'--',lw=2,color='grey')
T_keys = ['gistemp', 'hadcrut4', 'ncdc', 'berkeley', 'cowtanandway', 'hansen_1988_a_t', 'hansen_1988_b_t', 'hansen_1988_c_t']
first = 0
hansen_linestyle_idx = 0
hansen_linestyle = ['-','--',':']
hansen_labels = ['Scenario A', 'Scenario B', 'Scenario C']
for T_key in T_keys:
    if 'hansen' in T_key:
        plt.plot(filedata['year'],filedata[T_key], label=hansen_labels[hansen_linestyle_idx],
                 linestyle=hansen_linestyle[hansen_linestyle_idx],color='k',lw=2.5)
        hansen_linestyle_idx += 1
    else:
        if 'gistemp' in T_key: label = 'Observations'
        else: label = ''
        plt.plot(filedata['year'],filedata[T_key],label=label,color=cmap(0.7),lw=1.5)
plt.xlim([1957,2018])
plt.ylim([-0.3,1.35])
plt.grid(True)
plt.xlabel(r"year")
plt.ylabel(r"temperature anomaly [$\degree$C]")
plt.legend()

plt.subplot(2,1,2)
plt.pcolor(F,T,np.ma.masked_array(heatmap,heatmap==0.)/np.sum(heatmap),cmap=new_cmap)
plt.colorbar(label=r'probability of observations [bins of 0.05 W/m$^{2}$ and 0.05 $\degree$C]',
             orientation='horizontal',pad=0.2,aspect=32)

hansen_labels = ['a','b','c']
hansen_linestyle = ['-','--',':']
hansen_linestyle_idx = 0
for hansen_label in hansen_labels:
    plt.plot(filedata['hansen_1988_{0}_f'.format(hansen_label)],
             filedata['hansen_1988_{0}_t'.format(hansen_label)],
             linestyle=hansen_linestyle[hansen_linestyle_idx],color='k',lw=2)
    hansen_linestyle_idx += 1

plt.xlabel(r"radiative forcing with respect to 1958-1987 mean [W/m$^{2}$]")
plt.ylabel(r"temperature anomaly [$\degree$C]")
plt.clim([0,0.0075])
plt.ylim([-0.35,1.4])
plt.xlim([-0.8,3.0])
plt.grid(True)

plt.savefig("temperature_forcing.png",dpi=300,bbox_inches="tight")
```
